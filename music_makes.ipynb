```python
import numpy as np
import pandas as pd
```

```python
ds=open('music_data.txt').read().split()
```

```python
uniq=list(set(ds))
chars=dict((c,i) for i,c in enumerate(uniq))
rev_chars=dict((i,c) for i,c in enumerate(uniq))
```

```python
def get_one_hot_vec(txt):
    vec=np.zeros(len(uniq))
    vec[chars[txt]]=1
    return vec
```

```python
data=[]
for ic in range(len(ds)):
    data.append(get_one_hot_vec(ds[ic]))
data=np.array(data)
```

```python
seq_data=[]
seq_out=[]
max_len=3
for ix in range(len(ds)-max_len):
    seq_data.append(data[ix:ix+max_len])
    seq_out.append(data[ix+max_len])
seq_data=np.array(seq_data)
seq_out=np.array(seq_out)
```

```python
 print seq_data[0].shape, seq_out[0].shape
```

```python
import keras 
from keras.models import Sequential
from keras.layers import LSTM, Activation,Dense
from keras.optimizers import RMSprop
```

```python
model=Sequential()
model.add(LSTM(256,input_shape=(3,6179)))
model.add(Dense(6179))
model.add(Activation('softmax'))
model.summary()
optimizer = RMSprop(lr=0.01)
model.compile(loss='categorical_crossentropy',optimizer=optimizer,metrics=['accuracy'])
```

```python
model.fit(seq_data, seq_out, shuffle=True,
              batch_size=128,
              epochs=2, verbose=2)
```

```python
a=seq_data[0]
print a.shape
ao=[]
ao.append(a)
ao=np.array(ao)
ans=model.predict(ao)
idd=np.argmax(ans)
ans=np.zeros((6179))
ans[idd]=1
print ans
```

```python
no_notes=10000
gen_notes=[]
for i in range(3):
    gen_notes.append(data[i])
  
for ic in range(3,no_notes):  
    a=gen_notes[ic-3:ic]
    
    a=np.array(a)
    
    ao=[]
    ao.append(a)
    ao=np.array(ao)
    
    ans=model.predict(ao)
    idd=np.argmax(ans)
    ans=np.zeros((6179))
    ans[idd]=1
    gen_notes.append(ans)
    print ic
    
```

```python
gen_notes=np.array(gen_notes)
```

```python
melody=[]
for i in range(gen_notes.shape[0]):
    melody.append(rev_chars[np.argmax(gen_notes[i])])
```

```python
music=''
for i in range(gen_notes.shape[0]):
    if ((i+1)%8==0 ): 
        
        music=music+melody[i]+'\n'
    
    else:        
        music=music+melody[i]+' '
print music
```

```python
f=open('deep_jazz.txt','w')
f.write(music)
f.close()
```
